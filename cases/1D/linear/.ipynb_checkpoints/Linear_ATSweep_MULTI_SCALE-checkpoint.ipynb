```python
import torch
import torch.nn as nn
import numpy as np
import time
import scipy.io
torch.manual_seed(124)
np.random.seed(124)
from smt.sampling_methods import LHS
#torch.set_default_tensor_type(torch.DoubleTensor)
```

```python
def train(epoch):
    model.train()
    model.it = epoch
    def closure():
        optimizer.zero_grad()                                              
       # loss_pde = model.loss_pde(x_int)
       # loss_ic =  model.loss_ic(x_ic,u_ic)
       # loss_bd =  model.loss_bd(x_bdL,u_bdL) +\
       #            model.loss_bd(x_bdR,u_bdR)
        
        loss_pde = 0.0*model.loss_pde(x_int_ignore) + 0.01*model.loss_pde(x_int_covert) +\
                   100*model.loss_mm(x_int_memory,u_int_memory) + model.loss_pde(x_int_overt)
        
        loss_ic = 0.00*model.loss_ic(x_ic_ignore,u_ic_ignore) +\
                   0.01*model.loss_ic(x_ic_covert,u_ic_covert) +\
                    100*model.loss_ic(x_ic_memory,u_ic_memory) +\
                          model.loss_ic(x_ic_overt,u_ic_overt)
        
        #loss_bd = model.loss_bd(x_bdL,u_bdL_in,mark_bd) #+model.loss_bd(x_bdR,u_bdR_in,mark_bd) 
        loss_bd =  0.0*model.loss_bd(x_bdL_ignore,u_bdL_ignore) +\
                    0.01*model.loss_bd(x_bdL_covert,u_bdL_covert) +\
                     100*model.loss_bd(x_bdL_memory,u_bdL_memory) +\
                        model.loss_bd(x_bdL_overt,u_bdL_overt) +\
                   0.0*model.loss_bd(x_bdR_ignore,u_bdR_ignore) +\
                    0.01*model.loss_bd(x_bdR_covert,u_bdR_covert) +\
                     100*model.loss_bd(x_bdR_memory,u_bdR_memory) +\
                         model.loss_bd(x_bdR_overt,u_bdR_overt)
         
        loss = loss_pde + 10*loss_ic+10*loss_bd                                       

        print(f'epoch {epoch} loss_pde:{loss_pde:.8f}, loss_ic:{loss_ic:.8f}, loss_bd:{loss_bd:.8f}')
        #model.it = model.it + 1
        #outputfile = open('loss_history_burgers.dat','a+')
        #print(f'{model.it}  {loss_pde:.6f}  {loss_ic:.6f}  {loss:.6f}',file=outputfile)
        #outputfile.close() 
        loss.backward()
        return loss

    loss = optimizer.step(closure)
    loss_value = loss.item() if not isinstance(loss, float) else loss
    print(f'epoch {epoch}: loss {loss_value:.6f}')
    
def gradients(outputs, inputs):
    return torch.autograd.grad(outputs, inputs,grad_outputs=torch.ones_like(outputs), create_graph=True)

def to_numpy(input):
    if isinstance(input, torch.Tensor):
        return input.detach().cpu().numpy()
    elif isinstance(input, np.ndarray):
        return input
    else:
        raise TypeError('Unknown type of input, expected torch.Tensor or ' \
                        'np.ndarray, but got {}'.format(type(input)))


class DNN(nn.Module):
    def __init__(self):
        super(DNN, self).__init__()
        self.net = nn.Sequential()                                                  
        self.net.add_module('Linear_layer_1', nn.Linear(2, 120))                     
        self.net.add_module('Tanh_layer_1', nn.Tanh())                              

        for num in range(2, 5):                                                     
            self.net.add_module('Linear_layer_%d' % (num), nn.Linear(120, 120))       
            self.net.add_module('Tanh_layer_%d' % (num), nn.Tanh())                 
        self.net.add_module('Linear_layer_final', nn.Linear(120, 1))                 

    def forward(self, x):
        return self.net(x)

    def loss_pde(self, x):
        u = self.net(x)                                                
        du_g = gradients(u, x)[0]                                 
        u_t,u_x = du_g[:, :1],du_g[:,1:]
    #d = 0.01*abs(u_x)+1
        f = (((u_t +u_x))**2).mean() 
        if (len(x) == 0) :
            f = 0 
        #f = ((u_t +u_x)**2).mean() 
        return f
    
    def res_pde(self,x):
        y = self.net(x)
        Res = np.zeros((x.shape[0]))                                  
        u = y[:, 0:1]
        du_g = gradients(u, x)[0]                                  
        u_t,u_x = du_g[:, :1],du_g[:,1:]
        Res = (u_t - u_x)**2 
        return Res 
    
    def lambda_pde(self,x):
        y = self.net(x)
        Res = np.zeros((x.shape[0]))                                  
        
        u = y[:, 0:1]
        du_g = gradients(u, x)[0]                                  
        u_t,u_x = du_g[:, :1],du_g[:,1:]
        d = 0.1*(abs(u_x)-u_x) + 1
        return  d

    def loss_ic(self, x_ic, u_ic):
        y_ic = self.net(x_ic)                                                      
        u_ic_in = y_ic[:,0]
        loss_ics = ((u_ic_in - u_ic) ** 2).mean()
        if (len(x_ic) == 0):
            loss_ics = 0
        return loss_ics
    def loss_mm(self, x_ic, u_ic):
        y_ic = self.net(x_ic)                                                      
        u_ic_in = y_ic[:,0]
        loss_ics = ((u_ic_in - u_ic[:,0]) ** 2).mean()
        if (len(x_ic) == 0):
            loss_ics = 0
        return loss_ics
    def loss_bd(self, x,u_in):
        y = self.net(x)                                                      
        loss_bds = ((y[:,0] - u_in[:]) ** 2).mean()
        if (len(x) == 0):
            loss_bds = 0
        return loss_bds
    def loss_bd1(self, x):
        y = self.net(x)                                                      
        u = y[:,0:1]
        du_g = gradients(y,x)[0]
        u_x = du_g[:,1:]
        
        loss_bds = ((u + 1.0) ** 2 + (u_x)**2).mean()
        
        return loss_bds
def BD_perid(X,T,N):
    xL = np.zeros((N,2)) 
    xR = np.zeros((N,2)) 
    
    for i in range(N):
        t = np.random.rand()*T
        xL[i,0] = t
        xL[i,1] = -X 
        xR[i,0] = t
        xR[i,1] = X
    return xL,xR
def BD(X):
    u = np.zeros((X.shape[0])) 
    N = X.shape[0]
    for i in range(N):
        x = X[i,1]
        t = X[i,0]
        #u[i] =  np.sin(np.pi*(x-t))
        a = x-t
        if (a< -1):
            a =a+2
        elif(a > 1):
            a = a - 2    
        u[i] = Combination_wave(a)
    return u
def IC1(x):
    N = x.shape[0]
    u = np.zeros((N))                                                
    for i in range(N):
        u[i] = np.sin(np.pi*(x[i,1]))
    return u
def IC(X):
    N = X.shape[0]
    u = np.zeros((X.shape[0])) 
    for i in range(N):
        x = X[i,1]
        u[i] = Combination_wave(x)
        #u[i] = np.sin(np.pi*(x))
    return u
def Combination_wave(x):
    
    delta = 0.005
    a = 0.5
    z = -0.7
    alpha = 10
    beta = np.log(2)/(36*delta**2)
    
    if (x<-0.6 and x>=-0.8):
        u = 1/6*(G(x,beta,z-delta) + G(x,beta,z+delta)+4*G(x,beta,z))
    elif (x<-0.2 and x>=-0.4):
        u = 1
    elif (x>0 and x<0.2):
        u = 1-np.abs(10*(x-0.1))
    elif (x > 0.4 and x<0.6):
        u = 1/6*(F(x,alpha,a-delta)+F(x,alpha,a+delta)+4*F(x,alpha,a))
    else:
        u = 0
    return u

    

def G(x,beta,z):
    return np.exp(-beta*(x-z)**2)
def F(x,alpha,a):
    return np.sqrt(np.maximum(1-alpha**2*(x-a)**2,0))

def Mark(x):
    N =x.shape[0]
    mark = np.zeros((N,2)) 
    
    for i in range(N):
        if (x[i,1] < 0):
            mark[i,1] = 1.0
        elif ( x[i,0]< 0.1 and x[i,1]< 0.3 ):
            mark[i,1] = 1.0
        elif ( x[i,0]< 0.1 and x[i,1] > 0.3 ):
            mark[i,1] = 0.1
        else:
            mark[i,1] =  0.01 
    return mark

```

```python
def IgnoreAndAttention(X,xL,xR,tL,tR,num):
    A_covert    = []
    A_overt     = []
    A_memory    = []
    A_ignore    = []
    for i in range(num):
        x = X[i,1]
        t = X[i,0]
        if ((t - tL) < 0 ):
            A_memory.append(i)
        elif ((t - tL) >= 0 and (t- tR) <= 0 ):
            if ((x - xL) >= 0 and (x- xR) <= 0):
                A_overt.append(i)
            else:
                A_covert.append(i)
        else:
            A_ignore.append(i)
            
    x_ignore = np.zeros((len(A_ignore),2))
    x_covert = np.zeros((len(A_covert),2))
    x_overt = np.zeros((len(A_overt),2))
    x_memory = np.zeros((len(A_memory),2))
    for i in range( len(A_ignore)):
        x_ignore[i,:] = X[A_ignore[i],:]
    for i in range( len(A_memory)):
        x_memory[i,:] = X[A_memory[i],:]
        
    for i in range( len(A_covert)):
        x_covert[i,:] = X[A_covert[i],:]
    for i in range( len(A_overt)):
        x_overt[i,:] = X[A_overt[i],:]
    
    return x_memory, x_covert,x_overt,x_ignore
```

```python

device = torch.device('cuda')                                      
num_x = 500                                                        
num_t = 1000                                                        
num_ic = 1000                                           
num_int = 10000                                            
Time = 2.0
L = 1

AT_xL = -1.0
AT_xR = 1.0 
AT_tL = 1.0
AT_tR = 2.0

xlimits = np.array([[0.,Time],[-L, L]])  #interal
sampling = LHS(xlimits=xlimits)
x_int= sampling(num_int)
x_int_memory,x_int_covert,x_int_overt,x_int_ignore \
  = IgnoreAndAttention(x_int,AT_xL,AT_xR,AT_tL,AT_tR,num_int)

xlimits = np.array([[0.,0],[-L, L]])  #interal
sampling = LHS(xlimits=xlimits)
x_ic  = sampling(num_ic)
x_ic_memory,x_ic_covert,x_ic_overt,x_ic_ignore \
  = IgnoreAndAttention(x_ic,AT_xL,AT_xR,AT_tL,AT_tR,num_ic)
#u_ic = IC(x_ic)                 
u_ic_ignore = IC(x_ic_ignore)                 
u_ic_memory = IC(x_ic_memory)                 
u_ic_covert = IC(x_ic_covert)                 
u_ic_overt = IC(x_ic_overt)                 


xlimits = np.array([[0.,Time],[-L,-L]])  #interal
sampling = LHS(xlimits=xlimits)
x_bdL  = sampling(num_ic)

xlimits = np.array([[0.,Time],[L,L]])  #interal
sampling = LHS(xlimits=xlimits)
x_bdR  = sampling(num_ic)

x_bdL_memory,x_bdL_covert,x_bdL_overt,x_bdL_ignore \
= IgnoreAndAttention(x_bdL,AT_xL,AT_xR,AT_tL,AT_tR,num_t)
x_bdR_memory,x_bdR_covert,x_bdR_overt,x_bdR_ignore \
= IgnoreAndAttention(x_bdR,AT_xL,AT_xR,AT_tL,AT_tR,num_t)

u_bdL_memory =  BD(x_bdL_memory)                            
u_bdL_covert =  BD(x_bdL_covert)                            
u_bdL_overt =   BD(x_bdL_overt)                            
u_bdL_ignore =  BD(x_bdL_ignore)                            

u_bdR_memory =  BD(x_bdR_memory)                            
u_bdR_covert =  BD(x_bdR_covert)                            
u_bdR_overt =   BD(x_bdR_overt)                            
u_bdR_ignore =  BD(x_bdR_ignore)                            


x_ic_memory    = torch.tensor(x_ic_memory,                       dtype=torch.float32).to(device)
u_ic_memory    = torch.tensor(u_ic_memory,                       dtype=torch.float32).to(device)
x_int_memory   = torch.tensor(x_int_memory,                      dtype=torch.float32).to(device)
x_bdL_memory   = torch.tensor(x_bdL_memory,                      dtype=torch.float32).to(device)
x_bdR_memory   = torch.tensor(x_bdR_memory,                      dtype=torch.float32).to(device)
u_bdL_memory   = torch.tensor(u_bdL_memory,                      dtype=torch.float32).to(device)
u_bdR_memory   = torch.tensor(u_bdR_memory,                      dtype=torch.float32).to(device)
x_ic_covert    = torch.tensor(x_ic_covert,                       dtype=torch.float32).to(device)
u_ic_covert    = torch.tensor(u_ic_covert,                       dtype=torch.float32).to(device)
x_int_covert   = torch.tensor(x_int_covert, requires_grad=True,  dtype=torch.float32).to(device)
x_bdL_covert   = torch.tensor(x_bdL_covert, requires_grad=True,  dtype=torch.float32).to(device)
x_bdR_covert   = torch.tensor(x_bdR_covert, requires_grad=True,  dtype=torch.float32).to(device)
u_bdL_covert   = torch.tensor(u_bdL_covert, requires_grad=True,  dtype=torch.float32).to(device)
u_bdR_covert   = torch.tensor(u_bdR_covert, requires_grad=True,  dtype=torch.float32).to(device)
x_ic_overt    = torch.tensor(x_ic_overt,                         dtype=torch.float32).to(device)
u_ic_overt    = torch.tensor(u_ic_overt,                         dtype=torch.float32).to(device)
x_int_overt   = torch.tensor(x_int_overt, requires_grad=True,    dtype=torch.float32).to(device)
x_bdL_overt   = torch.tensor(x_bdL_overt,requires_grad=True,     dtype=torch.float32).to(device)
x_bdR_overt   = torch.tensor(x_bdR_overt, requires_grad=True,    dtype=torch.float32).to(device)
u_bdL_overt   = torch.tensor(u_bdL_overt, requires_grad=True,    dtype=torch.float32).to(device)
u_bdR_overt   = torch.tensor(u_bdR_overt, requires_grad=True,    dtype=torch.float32).to(device)
x_ic_ignore    = torch.tensor(x_ic_ignore,                       dtype=torch.float32).to(device)
u_ic_ignore    = torch.tensor(u_ic_ignore,                       dtype=torch.float32).to(device)
x_int_ignore   = torch.tensor(x_int_ignore, requires_grad=True,  dtype=torch.float32).to(device)
x_bdL_ignore   = torch.tensor(x_bdL_ignore, requires_grad=True,  dtype=torch.float32).to(device)
x_bdR_ignore   = torch.tensor(x_bdR_ignore, requires_grad=True,  dtype=torch.float32).to(device)
u_bdL_ignore   = torch.tensor(u_bdL_ignore, requires_grad=True,  dtype=torch.float32).to(device)
u_bdR_ignore   = torch.tensor(u_bdR_ignore, requires_grad=True,  dtype=torch.float32).to(device)

#u_bdL =  BD(x_bdL)                            
#u_bdR =  BD(x_bdR)                            

#x_ic    = torch.tensor(x_ic, dtype=torch.float32).to(device)
#u_ic    = torch.tensor(u_ic, dtype=torch.float32).to(device)
#x_int   = torch.tensor(x_int, requires_grad=True, dtype=torch.float32).to(device)
#x_bdL   = torch.tensor(x_bdL, requires_grad=True,  dtype=torch.float32).to(device)
#x_bdR   = torch.tensor(x_bdR, requires_grad=True,  dtype=torch.float32).to(device)
#u_bdL   = torch.tensor(u_bdL, requires_grad=True,  dtype=torch.float32).to(device)
#u_bdR   = torch.tensor(u_bdR, requires_grad=True,  dtype=torch.float32).to(device)

#model = DNN().to(device)


lr = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
print('Start training...')
```
