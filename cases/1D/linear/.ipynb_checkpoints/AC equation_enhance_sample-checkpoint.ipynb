# Burgers equation

This code is modified from the code https://github.com/alexpapados/Physics-Informed-Deep-Learning-Solid-and-Fluid-Mechanics

The paper about this work can ref https://www.researchgate.net/publication/359480166_Discontinuity_Computing_with_Physics-Informed_Neural_Network

```python
import torch
import torch.nn as nn
import numpy as np
import time
import scipy.io
torch.manual_seed(123)
np.random.seed(123)
from smt.sampling_methods import LHS
torch.set_default_tensor_type(torch.DoubleTensor)
```

```python
def train(epoch):
    model.train()
    model.it = epoch
    def closure():
        optimizer.zero_grad()                                              
        loss_pde = 0.001*model.loss_pde(x_int_ignore) + 0.1*model.loss_pde(x_int_covert) +\
                   100*model.loss_pde(x_int_memory) + model.loss_pde(x_int_overt)
        
        loss_ic = 0.001*model.loss_ic(x_ic_ignore,u_ic_ignore) +\
                   0.1*model.loss_ic(x_ic_covert,u_ic_covert) +\
                    100*model.loss_ic(x_ic_memory,u_ic_memory) +\
                          model.loss_ic(x_ic_overt,u_ic_overt)
        
        #loss_bd = model.loss_bd(x_bdL,u_bdL_in,mark_bd) #+model.loss_bd(x_bdR,u_bdR_in,mark_bd) 
        loss_bd =  0.001*model.loss_bd1(x_bdL_ignore) +\
                    0.1*model.loss_bd1(x_bdL_covert) +\
                     100*model.loss_bd1(x_bdL_memory) +\
                        model.loss_bd1(x_bdL_overt) +\
                   0.001*model.loss_bd1(x_bdR_ignore) +\
                    0.1*model.loss_bd1(x_bdR_covert) +\
                     100*model.loss_bd1(x_bdR_memory) +\
                        model.loss_bd1(x_bdR_overt)
        
         
        loss = loss_pde + 10*loss_ic+ 10*loss_bd                                       

        print(f'epoch {epoch} loss_pde:{loss_pde:.8f}, loss_ic:{loss_ic:.8f}, loss_bd:{loss_bd:.8f}')
        #model.it = model.it + 1
        #outputfile = open('loss_history_burgers.dat','a+')
        #print(f'{model.it}  {loss_pde:.6f}  {loss_ic:.6f}  {loss:.6f}',file=outputfile)
        #outputfile.close() 
        loss.backward()
        return loss

    loss = optimizer.step(closure)
    loss_value = loss.item() if not isinstance(loss, float) else loss
    print(f'epoch {epoch}: loss {loss_value:.6f}')
    
def gradients(outputs, inputs):
    return torch.autograd.grad(outputs, inputs,grad_outputs=torch.ones_like(outputs), create_graph=True)

def to_numpy(input):
    if isinstance(input, torch.Tensor):
        return input.detach().cpu().numpy()
    elif isinstance(input, np.ndarray):
        return input
    else:
        raise TypeError('Unknown type of input, expected torch.Tensor or ' \
                        'np.ndarray, but got {}'.format(type(input)))


class DNN(nn.Module):
    def __init__(self):
        super(DNN, self).__init__()
        self.net = nn.Sequential()                                                  
        self.net.add_module('Linear_layer_1', nn.Linear(2, 90))                     
        self.net.add_module('Tanh_layer_1', nn.Tanh())                              

        for num in range(2, 5):                                                     
            self.net.add_module('Linear_layer_%d' % (num), nn.Linear(90, 90))       
            self.net.add_module('Tanh_layer_%d' % (num), nn.Tanh())                 
        self.net.add_module('Linear_layer_final', nn.Linear(90, 1))                 

    def forward(self, x):
        return self.net(x)

    def loss_pde(self, x):
        y = self.net(x)                                                
        du_g = gradients(y, x)[0]                                 
        u_t,u_x = du_g[:, :1],du_g[:,1:]
        dux_g = gradients(u_x,x)[0]
        u_xx = dux_g[:,1:]
        f = ((u_t -0.001*u_xx+5*y**3-5*y)**2).mean() 
        if (len(x) == 0) :
            f = 0
        return f
    
    def res_pde(self,x):
        y = self.net(x)
        Res = np.zeros((x.shape[0]))                                  
        
        u = y[:, 0:1]
        du_g = gradients(u, x)[0]                                  
        u_t,u_x = du_g[:, :1],du_g[:,1:]
        Res = (u_t - u_x)**2 
        return Res 
    
    def lambda_pde(self,x):
        y = self.net(x)
        Res = np.zeros((x.shape[0]))                                  
        
        u = y[:, 0:1]
        du_g = gradients(u, x)[0]                                  
        u_t,u_x = du_g[:, :1],du_g[:,1:]
        d = 0.1*(abs(u_x)-u_x) + 1
        return  d

    def loss_ic(self, x_ic, u_ic):
        y_ic = self.net(x_ic)                                                      
        loss_ics = ((y_ic - u_ic) ** 2).mean()
        if (len(x_ic) == 0):
            loss_ics = 0
        return loss_ics
    def loss_bd(self, x,u_in):
        y = self.net(x)                                                      
        loss_bds = ((y - u_in) ** 2).mean()
        return loss_bds
    def loss_bd1(self, x):
        y = self.net(x)                                                      
       # uL = yL[:,0:1]
       # uR = yR[:,0:1]
     #   duL_g = gradients(uL,xL)[0]
     #   duR_g = gradients(uR,xR)[0]
        
     #   uL_x = duL_g[:,1:]
     #   uR_x = duR_g[:,1:]
        
        loss_bds = ((y + 1.0) ** 2).mean()
        
        if (len(x) == 0):
            loss_bds = 0
        return loss_bds
def BD_perid(X,T,N):
    xL = np.zeros((N,2)) 
    xR = np.zeros((N,2)) 
    
    for i in range(N):
        t = np.random.rand()*T
        xL[i,0] = t
        xL[i,1] = 0
        xR[i,0] = t
        xR[i,1] = X
    return xL,xR
def BD(X):
    u = np.zeros(X.shape[0]) 
    
    for i in range(X.shape[0]):
        x = X[i,1]
        t = X[i,0]
        u[i] =  np.sin(np.pi*(x-t))
    return u
def IC1(x):
    N = x.shape[0]
    u = np.zeros((N,1))                                                
    for i in range(N):
        u[i,0] = np.sin(np.pi*(x[i,1]))
    return u
def IC(X):
    N = len(X)
    u = np.zeros((X.shape[0]))                                                
    for i in range(N):
        x = X[i,1]
        u[i] =(x-1)**2*np.cos(np.pi*(x-1))
    return u

def Mark(x):
    N =x.shape[0]
    mark = np.zeros((N,2)) 
    
    for i in range(N):
        if (x[i,1] < 0):
            mark[i,1] = 1.0
        elif ( x[i,0]< 0.1 and x[i,1]< 0.3 ):
            mark[i,1] = 1.0
        elif ( x[i,0]< 0.1 and x[i,1] > 0.3 ):
            mark[i,1] = 0.1
        else:
            mark[i,1] =  0.01 
    return mark

```

```python
def IgnoreAndAttention(X,xL,xR,tL,tR,num):
    A_covert    = []
    A_overt     = []
    A_memory    = []
    A_ignore    = []
    for i in range(num):
        x = X[i,1]
        t = X[i,0]
        if ((t - tL) < 0 ):
            A_memory.append(i)
        elif ((t - tL) >= 0 and (t- tR) <= 0 ):
            if ((x - xL) >= 0 and (x- xR) <= 0):
                A_overt.append(i)
            else:
                A_covert.append(i)
        else:
            A_ignore.append(i)
            
    x_ignore = np.zeros((len(A_ignore),2))
    x_covert = np.zeros((len(A_covert),2))
    x_overt = np.zeros((len(A_overt),2))
    x_memory = np.zeros((len(A_memory),2))
    for i in range( len(A_ignore)):
        x_ignore[i,:] = X[A_ignore[i],:]
    for i in range( len(A_memory)):
        x_memory[i,:] = X[A_memory[i],:]
        
    for i in range( len(A_covert)):
        x_covert[i,:] = X[A_covert[i],:]
    for i in range( len(A_overt)):
        x_overt[i,:] = X[A_overt[i],:]
    
    return x_memory, x_covert,x_overt,x_ignore
```

```python

device = torch.device('cuda')                                      
num_x = 500                                                        
num_t = 100                                                        
num_ic = 100                                           
num_int = 20000                                            
Time = 1.0
L = 2

AT_xL = 0
AT_xR = 2.0
AT_tL = 0
AT_tR = 0.1

xlimits = np.array([[0.,Time],[0.0, L]])  #interal
sampling = LHS(xlimits=xlimits)
x_int= sampling(num_int)
x_int_memory,x_int_covert,x_int_overt,x_int_ignore = IgnoreAndAttention(x_int,AT_xL,AT_xR,AT_tL,AT_tR,num_int)

xlimits = np.array([[0.,0],[0.0, L]])  #interal
sampling = LHS(xlimits=xlimits)
x_ic  = sampling(num_ic)
x_ic_memory,x_ic_covert,x_ic_overt,x_ic_ignore = IgnoreAndAttention(x_ic,AT_xL,AT_xR,AT_tL,AT_tR,num_ic)

u_ic_ignore = IC(x_ic_ignore)                 
u_ic_memory = IC(x_ic_memory)                 
u_ic_covert = IC(x_ic_covert)                 
u_ic_overt = IC(x_ic_overt)                 

x_bdL,x_bdR = BD_perid(L,Time,num_t)                            
x_bdL_memory,x_bdL_covert,x_bdL_overt,x_bdL_ignore = IgnoreAndAttention(x_bdL,AT_xL,AT_xR,AT_tL,AT_tR,num_t)
x_bdR_memory,x_bdR_covert,x_bdR_overt,x_bdR_ignore = IgnoreAndAttention(x_bdR,AT_xL,AT_xR,AT_tL,AT_tR,num_t)

u_bdL_memory =  BD(x_bdL_memory)                            
u_bdL_covert =  BD(x_bdL_covert)                            
u_bdL_overt =   BD(x_bdL_overt)                            
u_bdL_ignore =  BD(x_bdL_ignore)                            

u_bdR_memory =  BD(x_bdR_memory)                            
u_bdR_covert =  BD(x_bdR_covert)                            
u_bdR_overt =   BD(x_bdR_overt)                            
u_bdR_ignore =  BD(x_bdR_ignore)                            

x_ic_memory    = torch.tensor(x_ic_memory, dtype=torch.float64).to(device)
u_ic_memory    = torch.tensor(u_ic_memory, dtype=torch.float64).to(device)
x_int_memory   = torch.tensor(x_int_memory, requires_grad=True, dtype=torch.float64).to(device)
x_bdL_memory   = torch.tensor(x_bdL_memory, dtype=torch.float64).to(device)
x_bdR_memory   = torch.tensor(x_bdR_memory, dtype=torch.float64).to(device)
u_bdL_memory   = torch.tensor(u_bdL_memory, dtype=torch.float64).to(device)
u_bdR_memory   = torch.tensor(u_bdR_memory, dtype=torch.float64).to(device)

x_ic_covert    = torch.tensor(x_ic_covert, dtype=torch.float64).to(device)
u_ic_covert    = torch.tensor(u_ic_covert, dtype=torch.float64).to(device)
x_int_covert   = torch.tensor(x_int_covert, requires_grad=True, dtype=torch.float64).to(device)
x_bdL_covert   = torch.tensor(x_bdL_covert, dtype=torch.float64).to(device)
x_bdR_covert   = torch.tensor(x_bdR_covert, dtype=torch.float64).to(device)
u_bdL_covert   = torch.tensor(u_bdL_covert, dtype=torch.float64).to(device)
u_bdR_covert   = torch.tensor(u_bdR_covert, dtype=torch.float64).to(device)

x_ic_overt    = torch.tensor(x_ic_overt, dtype=torch.float64).to(device)
u_ic_overt    = torch.tensor(u_ic_overt, dtype=torch.float64).to(device)
x_int_overt   = torch.tensor(x_int_overt, requires_grad=True, dtype=torch.float64).to(device)
x_bdL_overt   = torch.tensor(x_bdL_overt, dtype=torch.float64).to(device)
x_bdR_overt   = torch.tensor(x_bdR_overt, dtype=torch.float64).to(device)
u_bdL_overt   = torch.tensor(u_bdL_overt, dtype=torch.float64).to(device)
u_bdR_overt   = torch.tensor(u_bdR_overt, dtype=torch.float64).to(device)


x_ic_ignore    = torch.tensor(x_ic_ignore, dtype=torch.float64).to(device)
u_ic_ignore    = torch.tensor(u_ic_ignore, dtype=torch.float64).to(device)
x_int_ignore   = torch.tensor(x_int_ignore, requires_grad=True, dtype=torch.float64).to(device)
x_bdL_ignore   = torch.tensor(x_bdL_ignore, dtype=torch.float64).to(device)
x_bdR_ignore   = torch.tensor(x_bdR_ignore, dtype=torch.float64).to(device)
u_bdL_ignore   = torch.tensor(u_bdL_ignore, dtype=torch.float64).to(device)
u_bdR_ignore   = torch.tensor(u_bdR_ignore, dtype=torch.float64).to(device)

model = DNN().to(device)

lr = 0.001
#optimizer = torch.optim.Adam(model.parameters(), lr=lr)
print('Start training...')
```

```python
optimizer = torch.optim.LBFGS(model.parameters(),lr=0.1,max_iter=20)
epochs = 10000
tic = time.time()
for epoch in range(1, epochs+1):
    train(epoch)
    
toc = time.time()
```
