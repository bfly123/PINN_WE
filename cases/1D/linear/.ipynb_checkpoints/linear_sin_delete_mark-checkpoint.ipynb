```python
import torch
import torch.nn as nn
import numpy as np
import time
import scipy.io
torch.manual_seed(123)
np.random.seed(123)
from smt.sampling_methods import LHS
```

```python
def train(epoch):
    model.train()
    model.it = epoch
    def closure():
        optimizer.zero_grad()                                              
        loss_pde = model.loss_pde(x_int)                                   
        loss_ic = model.loss_ic(x_ic, u_ic)  
        loss_bd = model.loss_bd(x_bdL,u_bdL_in) #+model.loss_bd(x_bdR,u_bdR_in,mark_bd) 
        #loss_bd = model.loss_bd1(x_bdL,x_bdR,mark_bd) #+model.loss_bd(x_bdR,u_bdR_in,mark_bd) 
        loss = loss_pde + 10*(loss_ic+loss_bd)                                       

        print(f'epoch {epoch} loss_pde:{loss_pde:.8f}, loss_ic:{loss_ic:.8f}, loss_bd:{loss_bd:.8f}')
        #model.it = model.it + 1
        #outputfile = open('loss_history_burgers.dat','a+')
        #print(f'{model.it}  {loss_pde:.6f}  {loss_ic:.6f}  {loss:.6f}',file=outputfile)
        #outputfile.close() 
        loss.backward()
        return loss

    loss = optimizer.step(closure)
    loss_value = loss.item() if not isinstance(loss, float) else loss
    print(f'epoch {epoch}: loss {loss_value:.6f}')
    
def gradients(outputs, inputs):
    return torch.autograd.grad(outputs, inputs,grad_outputs=torch.ones_like(outputs), create_graph=True)

def to_numpy(input):
    if isinstance(input, torch.Tensor):
        return input.detach().cpu().numpy()
    elif isinstance(input, np.ndarray):
        return input
    else:
        raise TypeError('Unknown type of input, expected torch.Tensor or ' \
                        'np.ndarray, but got {}'.format(type(input)))


class DNN(nn.Module):
    def __init__(self):
        super(DNN, self).__init__()
        self.net = nn.Sequential()                                                  
        self.net.add_module('Linear_layer_1', nn.Linear(2, 20))                     
        self.net.add_module('Tanh_layer_1', nn.Tanh())                              

        for num in range(2, 5):                                                     
            self.net.add_module('Linear_layer_%d' % (num), nn.Linear(20, 20))       
            self.net.add_module('Tanh_layer_%d' % (num), nn.Tanh())                 
        self.net.add_module('Linear_layer_final', nn.Linear(20, 1))                 

    def forward(self, x):
        return self.net(x)

    def loss_pde(self, x):
        u = self.net(x)                                                
        du_g = gradients(u, x)[0]                                 
        u_t,u_x = du_g[:, :1],du_g[:,1:]
        f = (((u_t + u_x))**2).mean() 
         
        return f
    def res_pde(self,x):
        y = self.net(x)
        Res = np.zeros((x.shape[0]))                                  
        
        u = y[:, 0:1]
        du_g = gradients(u, x)[0]                                  
        u_t,u_x = du_g[:, :1],du_g[:,1:]
        Res = (u_t - u_x)**2 
        return Res 
    
    def lambda_pde(self,x):
        y = self.net(x)
        Res = np.zeros((x.shape[0]))                                  
        
        u = y[:, 0:1]
        du_g = gradients(u, x)[0]                                  
        u_t,u_x = du_g[:, :1],du_g[:,1:]
        d = 0.1*(abs(u_x)-u_x) + 1
        return  d

    def loss_ic(self, x_ic, u_ic):
        y_ic = self.net(x_ic)                                                      
        u_ic_nn = y_ic[:, 0]
        loss_ics = (((u_ic_nn - u_ic)) ** 2).mean()
        return loss_ics
    def loss_bd(self, x_bd,u_bd_in):
        y_bd = self.net(x_bd)                                                     
        u_bd = y_bd[:, 0]
        loss_bds = ((u_bd - u_bd_in) ** 2).mean()
        return loss_bds
    def loss_bd1(self, x_bdL,x_bdR,mark_bd):
        y_bdL = self.net(x_bdL)                                                      
        y_bdR = self.net(x_bdR)                                                      
        loss_bds = (((y_bdL[:,0] - y_bdR[:,0])) ** 2).mean()
        return loss_bds
def BD_perid(X,T,N):
    xL = np.zeros((N,2)) 
    xR = np.zeros((N,2)) 
    
    for i in range(N):
        t = (T*(i+1))/N
        xL[i,0] = t
        xL[i,1] = 0
        xR[i,0] = t
        xR[i,1] = X
    return xL,xR
def BD(X):
    u = np.zeros(X.shape[0]) 
    
    for i in range(X.shape[0]):
        x = X[i,1]
        t = X[i,0]
        u[i] =  np.sin(np.pi*(x-t))
    return u
def IC1(x):
    N = x.shape[0]
    u = np.zeros((N,1))                                                
    for i in range(N):
        u[i,0] = np.sin(np.pi*(x[i,1]))
    return u
def IC(x):
    N = len(x)
    u = np.zeros((x.shape[0]))                                                
    for i in range(N):
        u[i] = np.sin(np.pi*(x[i,1]))
    return u

def Mark(x):
    N =x.shape[0]
    mark = np.zeros((N)) 
    
    for i in range(N):
        if (x[i,1] < 0):
            mark[i] = 1.0
        elif ( x[i,1]< 20.0 and x[i,0] < 20):
            mark[i] = 1.0
        else:
            mark[i] =  0.01 
    return mark

```

```python

device = torch.device('cuda')                                      
num_x = 500                                                        
num_t = 500                                                        
num_i_train = 500                                           
num_ic =  500
num_f_train = 10000                                            
Time = 1
L = 2

x = np.linspace(0, L, num_x)                                   
t = np.linspace(0, Time, num_t)                                   
t_grid, x_grid = np.meshgrid(t, x)                             
T = t_grid.flatten()[:, None]                                  
X = x_grid.flatten()[:, None]                                  

xlimits = np.array([[0.,0],[0, L]])  #interal
sampling = LHS(xlimits=xlimits)
x_ic_train  = sampling(num_i_train)

u_ic_train = IC(x_ic_train)                 

xlimits = np.array([[0.,Time],[0, L]])  #interal
sampling = LHS(xlimits=xlimits)
x_int_train= sampling(num_f_train)

xlimits = np.array([[0.,Time],[0,0]])  #interal
sampling = LHS(xlimits=xlimits)
x_bdL  = sampling(num_ic)

xlimits = np.array([[0.,Time],[L,L]])  #interal
sampling = LHS(xlimits=xlimits)
x_bdR  = sampling(num_ic)

#x_bdL,x_bdR = BD_perid(L,Time,num_t)                            
u_bdL_in = BD(x_bdL)                            
u_bdR_in = BD(x_bdR)                            
x_ic = torch.tensor(x_ic_train, dtype=torch.float32).to(device)
x_int = torch.tensor(x_int_train, requires_grad=True, dtype=torch.float32).to(device)
u_ic = torch.tensor(u_ic_train, dtype=torch.float32).to(device)
x_bdL = torch.tensor(x_bdL, dtype=torch.float32).to(device)
x_bdR = torch.tensor(x_bdR, dtype=torch.float32).to(device)
u_bdL_in = torch.tensor(u_bdL_in, dtype=torch.float32).to(device)
u_bdR_in = torch.tensor(u_bdR_in, dtype=torch.float32).to(device)

model = DNN().to(device)

lr = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
print('Start training...')
```

```python
optimizer = torch.optim.LBFGS(model.parameters(),lr=1.0,max_iter=20)
epochs = 1000
tic = time.time()
for epoch in range(1, epochs+1):
    train(epoch)
    
toc = time.time()
```
