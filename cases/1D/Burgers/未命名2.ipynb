```python
import matplotlib.pyplot as plt

# x轴数据
x = [1, 2, 3, 4, 5]

# y轴数据
y = [2, 3, 5, 1, 4]

# 每个点的偏差范围
y_err = [0.5, 0.3, 0.7, 0.4, 0.6]

# 绘制折线图
plt.plot(x, y, color='k', linewidth=2)

# 绘制每个点的偏差范围
for i in range(len(x)):
    plt.fill_between([x[i]-0.1, x[i]+0.1], [y[i]-y_err[i], y[i]-y_err[i]], [y[i]+y_err[i], y[i]+y_err[i]], color='gray', alpha=0.2)

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# 生成一些随机数据
x = np.linspace(10, 100, num=100)
y = np.random.rand(100) * 100

# 绘制带有对数坐标轴的折线图
fig, ax = plt.subplots()
ax.set_xscale('log2')
ax.set_yscale('log2')
ax.plot(x, y)

# 定义格式化函数
def format_func(value, tick_number):
    return str(round(value/10, 1))

# 设置坐标轴上的格式化函数
ax.xaxis.set_major_formatter(FuncFormatter(format_func))
ax.yaxis.set_major_formatter(FuncFormatter(format_func))

# 设置图表标题和坐标轴标签
ax.set_title('Example of Logarithmic Axis')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# 显示图表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# 生成一些随机数据
x = np.linspace(10, 160, num=100)
y = np.random.rand(100) * 100

# 设置函数
def f(x):
    return np.log2(x)

# 绘制带有函数坐标轴的折线图
fig, ax = plt.subplots()
ax.plot(f(x), y)

# 定义格式化函数
def format_func(value, tick_number):
    return str(round(value, 1))

# 设置坐标轴上的格式化函数
ax.xaxis.set_major_formatter(FuncFormatter(format_func))

# 设置坐标轴的刻度
ticks = np.arange(1, 160)
ax.set_xticks(ticks)
ax.set_xticklabels(ticks)

# 设置图表标题和坐标轴标签
ax.set_title('Example of Function Axis')
ax.set_xlabel('$f(x) = x^2$')
ax.set_ylabel('Y-axis')

# 显示图表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator, FixedFormatter

# 生成一些随机数据
x = [8,16,32,64,128]
y = np.random.rand(5,2)

# 绘制折线图
fig, ax = plt.subplots()
ax.plot(x, y[:,0],label=r'$labe^{\infty}$')
ax.plot(x, y[:,1],label='label 2')
for i in range(len(x)):
    plt.fill_between([x[i]-0.01*x[i], x[i]+0.01*x[i]], [y[i,0]-0.1, y[i,0]-0.1], [y[i,0]+0.1, y[i,0]+0.1], color='blue', alpha=0.2)
# 设置 x 轴的刻度位置和标签
xticks = [8, 16, 32, 64, 128]
ax.set_xscale('log', base=2)
ax.set_yscale('log')
ax.set_xticks(xticks)
ax.xaxis.set_major_locator(FixedLocator(xticks))
ax.xaxis.set_major_formatter(FixedFormatter(xticks))
ax.legend(loc='upper right')
# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个空的图形和坐标系
fig, ax = plt.subplots()

# 设置 x 轴的范围和标签
ax.set_xlim([0, 150])
ax.set_xticks([8, 16, 32, 64, 128])
ax.set_xticklabels(['8', '16', '32', '64', '128'])

# 将刻度标签居中对齐
ax.tick_params(axis='x', pad=10)

# 隐藏顶部和右侧的坐标轴线
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# 设置 x 轴的坐标刻度为等距
x = np.linspace(0, 150, num=6)
ax.set_xticks(x)

# 显示图表
plt.show()
```
